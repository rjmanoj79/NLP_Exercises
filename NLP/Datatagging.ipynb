```python
df1=pd.read_excel("D:\AI_ML_Study\DataCatalog\Dataset2.xlsx",encoding = "ISO-8859-1")
df1.head()
```

#### Data Pre-Processing 

#### Remove Comma from the value

```python
df1['Employee_Name']=df1['Employee_Name'].str.replace(',','')
df1.head()
```

#### Split the value as First Name and Last Name

```python
new = df1["Employee_Name"].str.split(" ", n = 1, expand = True)
df1["First Name"]= new[0] 
df1["Last Name"]= new[1] 
df1.head()
```

#### Dropping Null values & Describing the value

```python
df1.dropna(inplace=True)
```

```python
df1.info()
```

```python
df1['Employee_Name'].describe()
```

#### POS Tagging 

```python
df1['FirstName_POSTags'] = pos_tag_sents(df1['First Name'].apply(word_tokenize).tolist())
df1.head()
```

```python
df1['LastName_POSTags'] = pos_tag_sents(df1['Last Name'].apply(word_tokenize).tolist())
df1.head()
```

```python
column_names = ["Attr_type", "Attr_spl_char", "Attr_col_name","Attr_count","Attr_unique"]
name_prof=pd.DataFrame(columns=column_names)
name_prof.columns
```

#### Count & Unique Values for Employee Name

```python
count=df1['Employee_Name'].count()
name_prof['Attr_count']=[count]
```

```python
uniq=df1['Employee_Name'].nunique()
name_prof['Attr_unique']=[uniq]
```

#### Statistics for the column having string

```python
type = df1.dtypes['Employee_Name']
if type==np.object:
    name_prof['Attr_type'] =['Object']
else:
    name_prof['Attr_type'] =['NA']
```

#### Check Values have Special Characters

```python
regex = re.compile('[@_!#$%^&*()<>?/\|}{~:]')
list_name=df1["First Name"]
for item in list_name:
    if(regex.search(item) == None):
        name_prof['Attr_spl_char']= ['No']
    else: 
        name_prof['Attr_spl_char']= ['Yes']
        break
```
